## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#Load data from json object into a dataframe, inspect info about columns
world_bank_df = pd.read_json('data/world_bank_projects.json')
world_bank_df.info()
```

```python
world_bank_df.head()
```

### Question 1###

```python
#1.Find the 10 countries with the most projects
#First, group data by country name, then aggregate by count over project name column
num_projects_by_country = world_bank_df.groupby('countryname')['project_name'].count()

#Sort the values in the series resulting from grouping by country and aggregating by number of projects, return top 10
num_projects_by_country.sort_values(ascending=False).head(10)
```

The 10 countries with the most projects are listed above. Although Africa is a continent, we can conclude from the fact that Morocco and Mozambique are included as separate countries that 'Africa' in this data refers to borrowers within Africa that are not tied to a unique country.

### Question 2: Find the top 10 major project themes using column 'mjtheme_namecode' ###

Based on the string format of the World Bank json data displayed earlier in the file, we know that 'mjtheme_namecode' is a nested column that includes the name of the project theme as well as its numerical encoding. Therefore, we need to normalize this column to convert it into a separate table.

```python
#First load the json as a string

world_bank_str = json.load((open('data/world_bank_projects.json')))

#Convert the nested column 'mjtheme_namecode' into its own table using json_normalize

project_theme_table = json_normalize(world_bank_str, 'mjtheme_namecode')
project_theme_table.head(15)
```

```python
#Group project theme data by the project theme code (we can't group by name because some names are missing) and count the number of entries corresponding to each code
num_projects_by_code = project_theme_table.groupby('code')['code'].count()
num_projects_by_code.sort_values(ascending=False).head(10)
```

To get a table that lists the theme name that corresponds to each numerical code, we can use the 'project_theme_table' created above. We will get rid of rows with missing names first, and then get rid of duplicates.

```python
code_to_names = project_theme_table[project_theme_table.name != ''].drop_duplicates()
code_to_names
```

Using this table along with the Series 'num_projects_by_code' created earlier, which lists the codes with the most projects, we conclude that the top 10 major project themes are:

1. Environment and natural resources management
2. Rural development
3. Human development
4. Public sector governance
5. Social protection and risk management
6. Financial and private sector development
7. Social dev/gender/inclusion
8. Trade and integration
9. Urban development
10. Economic management


### Question 3: Create a dataframe with the missing names from the normalized 'mjtheme_namecode' column filled in###

<p>The tabular format of the nested 'mjtheme_namecode' column is the dataframe 'project_theme_table', and it shares the 'code' column with the dataframe 'code_to_names', which maps code numbers to code names. We can merge these two dataframes together on the shared 'code' column. Since this column is common to both dataframes, each code value in 'project_theme_table' will match with a unique value in 'code_to_names'.</p>
<p>The result of the merge is a new dataframe where each row includes the entry in the 'name' column from 'project_theme_table' as well as the entry in the 'name' column from the corresponding row (matched by code value) in 'code_to_names'.</p> 
<p>Since all the records of the 'name' column from 'code_to_names' are filled in, the merged dataframe will have a 'name' column with no missing values. We can then drop the 'name' column with missing values from the merged dataframe.</p>

```python
new_table = pd.merge(project_theme_table, code_to_names, on='code', suffixes=['_project_theme','_code_to_names'])
new_table.drop('name_project_theme', axis=1, inplace=True)
new_table
```
